IMPORTING LIBRARIES





```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING THE DATASET

```python
#Loading dataset and reading CSV file
paitent_data = pd.read_csv('/content/OCD Patient Dataset_ Demographics & Clinical Data.csv')
```

```python
paitent_data.columns=paitent_data.columns.str.strip()
```

SHOWING DATAHEAD

```python
paitent_data.head()
```

```python
#Drop duplicates rows across all columns
paitent_data = paitent_data.drop_duplicates()
```

```python
#Drop column: 'Patient ID'
paitent_data = paitent_data.drop(columns=['Patient ID'])
```

```python
#Replace missing values with 0 in columns: 'Patient ID','Age' and other columns
paitent_data = paitent_data.fillna({'Duration of Symptoms (months)': 0, 'Y-BOCS Score (Obsessions)': 0, 'Y-BOCS Score (Compulsions)': 0})
```

```python
#Printing the dimensions of the Dataframe
print (paitent_data.shape)
```

```python
#this will show the number of null or blank spaces in the data
paitent_data.isnull().sum()
```

```python
#Find the mode of 'Privious Diagnoses' column
previous_diagnoses_mode = paitent_data['Previous Diagnoses'].mode()[0]

# Find the mode of 'Medications' column
medications_mode = paitent_data['Medications'].mode()[0]

# Fill missing values with the mode
paitent_data['Previous Diagnoses'].fillna(previous_diagnoses_mode, inplace=True)
paitent_data['Medications'].fillna(medications_mode, inplace=True)
paitent_data
```

```python
#Removing Leading WhiteSpace in Columns
paitent_data['Education Level'] = paitent_data['Education Level'].str.lstrip()
paitent_data['OCD Diagnosis Date'] = paitent_data['OCD Diagnosis Date'].str.lstrip()
paitent_data['Previous Diagnoses'] = paitent_data['Previous Diagnoses'].str.lstrip()
paitent_data['Family History of OCD'] = paitent_data['Family History of OCD'].str.lstrip()
paitent_data['Obsession Type'] = paitent_data['Obsession Type'].str.lstrip()
paitent_data['Compulsion Type'] = paitent_data['Compulsion Type'].str.lstrip()
paitent_data['Depression Diagnosis'] = paitent_data['Depression Diagnosis'].str.lstrip()
paitent_data['Anxiety Diagnosis'] = paitent_data['Anxiety Diagnosis'].str.lstrip()
paitent_data['Medications'] = paitent_data['Medications'].str.lstrip()
```

```python
#replacing the value in the 'gender' column with 1 if the value is 'Female' and 0 otherwise

paitent_data['Gender'] = paitent_data['Gender'].apply(lambda x : 1 if x.lower() == 'female' else 0)
paitent_data['Gender']
```

```python
# define the maping

previous_diagnoses_mapping = {
    'MMD' : 0,
    'GAD' : 1,
    'Panic Disorder' : 2,
    'PTSD' : 3
}

# Apply the mapping to the 'Previous Diagnoses' column
paitent_data['Previous Diagnoses'] = paitent_data['Previous Diagnoses'].map(previous_diagnoses_mapping)
```

```python
# Define the mapping
ethnicity_mapping = {
    'Caucasian': 0,
    'Asian': 1,
    'African': 2,
    'Hispanic': 3
}

# Apply the mapping to the 'Ethnicity' column
paitent_data['Ethnicity'] = paitent_data['Ethnicity'].map(ethnicity_mapping)
```

```python
# Define the mapping
marital_status_mapping = {
    'Single': 0,
    'Married': 1,
    'Divorced': 2
}

# Apply the mapping to the 'Marital Status' column
paitent_data['Marital Status'] = paitent_data['Marital Status'].map(marital_status_mapping)
```

CREATING THE BAR PLOT OF 'OBSESSION TYPE BY GENDER'

```python
plt.figure(figsize=(7,8))
sns.set(style="whitegrid")

#create the count plot with specificed palette
plot = sns.countplot(data=paitent_data, x='Gender', hue='Obsession Type', palette= 'Oranges' )

#set the background color
sns.set(rc = {'axes.facecolor': 'lightblue', 'figure.facecolor': 'lightblue'})
plt.title('Obsession Type by Gender')

#add data labels

for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha = 'center', va = 'center',
                  xytext = (0, 10),
                  textcoords = 'offset points')

# Show the plot
plt.show()
```

CREATING THE SWARMPLOT OF 'OBSESSIO AND COMPULSION TYPE WITH REFRENCE TO AGE'

```python
#set the background color
sns.set(rc = {'axes.facecolor':'lightblue', 'figure.facecolor': 'lightblue'})

#creat a scatterplot
plt.figure(figsize=(10, 6))
sns.swarmplot(data=paitent_data, x='Obsession Type', y='Age', hue='Compulsion Type', palette='icefire')

#add a title,axis and legend
plt.title('obsession and compulsion type')
plt.xlabel('Obsession Type')
plt.ylabel('Age')
plt.legend(title='Compulsion Type', title_fontsize='13', loc='upper right')
plt.show()

```

DIPRESSION AND ANXIETY DIAGNOSES BY GENDER :BAR PLOT

```python
paitent_data = {
    'Gender' : ['Male', 'Female', 'Male', 'Female','Male', 'Female'],
    'Diagnosis' : ['Dipression', 'Dipression', 'Anxity', 'Anxity', 'Dipression', 'Anxity'],
    'Count' :[20, 30, 25, 35, 18, 22]
    }
# create a Dataframe From the smaple data
df_sample = pd.DataFrame(paitent_data)

# Set the background color
sns.set(rc={'axes.facecolor':'grey', 'figure.facecolor':'grey'})

#creat the grouped bar plot
plt.figure(figsize=(10, 6))
plot =  sns.barplot(data=df_sample, x='Gender', y='Count', hue='Diagnosis', palette='pastel')
plt.title(' Distribution of Dipression and Anxity Diagnosis by Gender')

```

CREATING SCATTERPLOT OF 'AGE VS MEDICATION'

```python
plt.figure(figsize=(10, 6))
sns.stripplot(
    data=paitent_data,
    x='Medications',
    y='Age',
    jitter=True,
    hue='Age',
    palette='icefire'
)
plt.title('paitent Age vs Medications')
plt.xlabel('Medications')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

CREATING THE COUNTPLOT OF DISTRIBUTION OF ANXIETY DIAGNOSES BY EDUCATION LEVEL'

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=paitent_data, x='Education Level', hue='Anxiety Diagnosis', palette='pastel')
plt.title('Count of Anxiety Diagnosis by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
```
